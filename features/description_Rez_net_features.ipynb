```python
import sys  
sys.path.append("../../")
from LIB.utils import * 
from sklearn.decomposition  import  PCA
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
test=pd.read_csv("../Data/raw_data/test.csv")
data_characterization(test)
train=pd.read_csv("../Data/raw_data/train.csv")
data_characterization(train)
```

```python
data=pd.concat([train,test])
```

```python
data=data[["ID","en_US_description"]]
```

```python
vectorizer = TfidfVectorizer()
corpus=data.en_US_description.values.tolist()
X = vectorizer.fit(corpus)
```

```python
test_tf_idf=X.transform(corpus)
test_tf_idf.shape
```

```python
test_tf_idf=X.transform(corpus)
test_DF=pd.DataFrame()
pca_test=PCA(n_components=20)
test_tf_idf_array=pca_test.fit_transform(test_tf_idf.toarray())
for i in range(test_tf_idf_array.shape[1]) :
    test_DF["type_tf_idf_pca_"+str(i)]=test_tf_idf_array[:,i]
```

```python
test_DF["ID"]=data["ID"].tolist()
```

```python
test_DF.to_csv("Tf_idf_20.csv",index=False)
```
