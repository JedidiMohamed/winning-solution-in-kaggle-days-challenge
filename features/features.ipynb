```python
test=pd.read_csv("../Data/raw_data/test.csv")
data_characterization(test)
train=pd.read_csv("../Data/raw_data/train.csv")
data_characterization(train)
```

```python
map_categorical_feautres(Data=[train,test],columns=["product_type","product_gender","macro_function","function",
                                                   "sub_function","aesthetic_sub_line","macro_material"])
```

```python

train["target_log"]=np.log1p(train.target)
col_to_aggr=[["product_type"],["product_gender"],["macro_function"],["function"],["sub_function"],["aesthetic_sub_line"],
            [ "macro_material"]
 
             ]
train_df,test_df=aggr_func(train=train,test=test,vars_to_agg=col_to_aggr,func=["mean","min","std"],
         vars_be_agg="target_log",fillnan=True,bagging=False,ID="ride_id" ,n_folds=6,shuffle=True)
```

```python
sales=pd.read_csv("../Data/raw_data/sales.csv")
sales.head()
```

```python
count_sales_first_days=sales.groupby("sku_hash").sales_quantity.sum().rename("Sales_first_7_days").reset_index()
train_df=train_df.merge(count_sales_first_days,how="left",on="sku_hash")
test_df=test_df.merge(count_sales_first_days,how="left",on="sku_hash")
```

```python
first_month=sales.groupby("sku_hash").Month_transaction.min().rename("first_month_sales").reset_index()
train_df=train_df.merge(first_month,how="left",on="sku_hash")
test_df=test_df.merge(first_month,how="left",on="sku_hash")
```

```python
last_days=sales[sales.Date.isin(["Day_7","Day_6"])]
count_sales_last_days=last_days.groupby("sku_hash").sales_quantity.sum().rename("Sales_last_days").reset_index()
train_df=train_df.merge(count_sales_last_days,how="left",on="sku_hash")
test_df=test_df.merge(count_sales_last_days,how="left",on="sku_hash")
```
